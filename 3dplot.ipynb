```python
import requests, re, csv, sys

import pandas as pd
from bs4 import BeautifulSoup
import bs4 as bs
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D


import warnings
warnings.filterwarnings("ignore")

gourl = "india"
basepath = "C:"
codepath="python"
function = "temples"
N="\\"
namefile = "googledata"

inputcsvfilename = ("{}{}{}{}".format(namefile,"_",gourl,"_temples.csv"))

googletempledata = ("{}{}{}{}{}{}{}{}{}{}{}".format(basepath,N,codepath,N,function,N,"data",N,"STATEWISE_TEMPLES",N,inputcsvfilename))

temple_data = pd.read_csv(googletempledata)

#temple_data['State']=temple_data['templeName'].str.split(',',expand=True)[1]

def pltfig(parameter):

    plt.figure(figsize=(15,6))
    sns.countplot(parameter,
                  x = parameter,  
                  data = temple_data, 
                  palette="Set1"
                
                  )
    plt.xticks(rotation = 270)
    
    plt.show()

     


#pltfig('DistanceFromMumbai(Km)')

def templesview(v,g,ln):
   
    tv = temple_data.sort_values(by=[g, v],
    ignore_index=True).loc[:,[g,v]].head(ln)
    return tv


v = "Distance From Bengaluru(Km)"
g = 'Temple Name'
la = len(temple_data)
sn = 25
if(la < sn):
    ln = la
else:
    ln = sn
bangaloredistance = templesview(v,g,ln)




def pltleg(x,y,g,v):
    #plt.plot(x, label = g)
    plt.plot(y, label = v)
    
    plt.legend(bbox_to_anchor =(.75, 1.15), ncol = 3)
    plt.xticks(rotation = 90)
    plt.show()


pltfig('State')
#pltfig(v)
bld = (bangaloredistance[v])
x= (bangaloredistance[g])
pltleg(x,bld,g,v)

stc = temple_data.State.value_counts()

stu = temple_data.State.unique()

pltfig(stc)



plt.plot(stc, label = 'States wise Temples Count')
plt.legend(bbox_to_anchor =(.75, 1.15), ncol = 3)
plt.xticks(rotation = 90)
plt.show()


```
